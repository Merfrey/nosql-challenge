### 1. Which establishments have a hygiene score equal to 20?

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query2 = db.establishments.find({'LocalAuthorityName' : { '$regex': 'London'}, 'RatingValue': {'$gte': 4}})
results2 = [x for x in query2]
#query = {'LocalAuthorityName': {'Your search term here'}, 'RatingValue': {'$gte': <number you want here>}}

# Use count_documents to display the number of documents in the result

query2Counter = db.establishments.count_documents({'LocalAuthorityName' : { '$regex': 'London'}, 'RatingValue': {'$gte': 4}})
pprint(f"The number of establishments in London with a RatingValue greater than or equal to 4 is {query2Counter}")
pprint(f"+++++++++++++++++++++++++++++++++++++++++++++")

# Display the first document in the results using pprint
pprint(establishments.find_one({'LocalAuthorityName' : { '$regex': 'London'}, 'RatingValue': {'$gte': 4}}))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score


degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

query3 = {'geocode.latitude':{'$lte': latitude + degree_search, '$gte':latitude - degree_search},
        'geocode.longitude':{'$lte': longitude + degree_search, '$gte':longitude - degree_search},
        'RatingValue': 5}

sort = [('scores.Hygine', 1)]

limit = 5

results3 = establishments.find(query3).sort(sort).limit(limit)

# Print the results
pprint(list(establishments.find(query3).sort(sort).limit(limit)))
```

```python
# Convert result to Pandas DataFrame
results3DF = pd.DataFrame(results3)
```
